```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset into a pandas dataframe
#df = pd.read_csv('data_yield copy 2.csv')

df = pd.read_csv('new_data_full.csv')
df = df.loc[:, ~df.columns.str.contains('9')]
df = df.drop(columns=['Division','Year']).dropna()
df['Soil Type'] = df['Soil Type'].astype('category')
df['Soil Type'] = df['Soil Type'].cat.codes

# Create a box plot of the selected columns using Seaborn
sns.set(style='ticks', font_scale=1.2)
ax = sns.boxplot(data=df, palette='Set3')
ax.set_ylabel('Value', fontsize=12)
ax.set_xlabel('Features', fontsize=10)
ax.set_title('Environmental Data Normal', fontsize=14)

# Rotate the x-axis labels
plt.xticks(rotation=45, ha='right',fontsize=10)
plt.tight_layout()

plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

# Standardize the dataframe using Scikit-learn's StandardScaler
scaler = StandardScaler()
standardized_df = scaler.fit_transform(df)

# Convert the standardized numpy array back to a pandas dataframe
standardized_df = pd.DataFrame(standardized_df, columns=df.columns)

# Create a box plot of the selected columns using Seaborn
sns.set(style='ticks', font_scale=1.2)
ax = sns.boxplot(data=standardized_df, palette='Set3')
ax.set_ylabel('Value', fontsize=12)
ax.set_xlabel('Features', fontsize=10)
ax.set_title('Environmental Data Standardized', fontsize=14)

# Rotate the x-axis labels
plt.xticks(rotation=45, ha='right',fontsize=10)
plt.tight_layout()

plt.show()
```
