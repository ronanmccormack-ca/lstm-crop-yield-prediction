```python
# Import the libraries
import pandas as pd
from datetime import datetime
import os
import re
import seaborn as sns
import matplotlib.pyplot as plt

# Define the file pattern to look for
file_pattern = 'TH'

# Define the allowed file extensions
allowed_extensions = ['.xls', '.xlsx']

# Create an empty list to store the matching file names
file_list = []

# Loop through all files in the directory
for filename in os.listdir():
    # Check if the filename contains the specified pattern
    if file_pattern in filename:
        # Check if the file has an allowed extension
        if any(ext in filename for ext in allowed_extensions):
            # Add the filename to the list
            file_list.append(filename)

# Create empty DataFrame to store merged data
merged_df = pd.DataFrame()

# Loop through each file
for i in file_list:
    # Read in file as DataFrame
    df = pd.read_excel(i)
    
    # Append DataFrame to merged_df
    merged_df = pd.concat([merged_df, df])

# Loop through each file
for i in file_list:
    # Read in file as DataFrame
    df = pd.read_excel(i)
    
    # Append DataFrame to merged_df
    merged_df = pd.concat([merged_df, df])
    
# Reset index of merged_df
merged_df.reset_index(inplace=True, drop=True)

# Change the date format from xx/xx/xx to xx-xx-xx
merged_df['Date'] = pd.to_datetime(merged_df['Date'], errors='coerce')
merged_df['Date'] = merged_df['Date'].dt.strftime('%m-%d-%Y %I:%M %p')
merged_df['Date'] = pd.to_datetime(merged_df['Date']).dt.date

# Defining the function to extract the temperature element from the Formatted Value column
def extract_temperature(x):
    try:
        temperature = re.search(r'\d+\.\d+', x.split('@')[-1]).group()
    except (IndexError, AttributeError):
        temperature = None
    return temperature

# Example usage
merged_df['Formatted Value'] = merged_df['Formatted Value'].apply(extract_temperature)

# Creating a new DataFrame to hold relevant data
df = merged_df[['SensorID','Date','Value','Formatted Value']]

# Renaming the columns
df = df.rename(columns={'Value': 'Humidity', 'Formatted Value': 'Temperature'})

# Convert Date column to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Convert Temperature column to float
df['Temperature'] = df['Temperature'].astype(float)

# Convert SensorID column to int
df['SensorID'] = df['SensorID'].astype(int)

# Grouping by Date and getting the mean for each data point
df = df.groupby('Date').agg({'Humidity': 'mean', 'Temperature': 'mean'})

# Move the 'Date' index back to a regular column
df = df.reset_index()

# Export to CSV
df.to_csv('TH DATA.csv',index=False)

# Set Seaborn Style
sns.set_style("darkgrid")
sns.set_palette("Set2")

# create the plot
sns.lineplot(x="Date", y="Temperature", data=df, label="Temperature")
sns.lineplot(x="Date", y="Humidity", data=df, label="Humidity")

# set the title and axes labels
plt.title("Temperature and Humidity Over Time")
plt.xlabel("Date")
plt.ylabel("Value")

# Save the plot to a file
plt.savefig('TH DATA.png')
```
